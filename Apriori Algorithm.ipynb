```python
#input file name will be: dataset_1.csv, dataset_2.csv
#Here are an example input file parser:
import csv
import numpy as np

########  A List是否包含 B List 回傳共包含ｘ個 ########
def Contain(A,B):
    count = 0
    for index in A:
        i=0
        j=0
        while i<len(index):
            if index[i] == B[j]:
                i+=1
                j+=1
                if j==len(B):
                    count += 1
                    break
            else:
                i+=1
    return count

########  Sorting  ########
def Sorting(List1,List2):
    for i in range(len(List1)):
        for j in range(len(List1)):
            if j+1 < len(List1):
                for k in range(min(len(List1[j]),len(List1[j+1]))):
                    if int(List1[j][k]) > int(List1[j+1][k]):
                        temp = List1[j+1]
                        List1[j+1] = List1[j]
                        List1[j] = temp
                        temp = List2[j+1]
                        List2[j+1] = List2[j]
                        List2[j] = temp
                        break
                    if int(List1[j][k]) < int(List1[j+1][k]):
                        break
########  主要程式  ########
def HW2(support,input_file,output_file):
    db = []
    answer_list=[]
    answer_support=[]

    with open(input_file, newline='') as csvfile:
        rows = csv.reader(csvfile)
        for row in rows:
            db.append(row)
    ########  find maximum value  ########
    Max_value=0
    for index1 in db:
        for index2 in index1:
            if int(index2)>=Max_value:
                Max_value = int(index2)
    ########  包含一個元素的sublist  ########
    for i in range(Max_value+1):
        count = 0
        for index1 in db:
            for index2 in index1:
                if int(index2) == i:
                    count+=1
                    break
        if count >= support:
            answer_list.append([str(i)])
            answer_support.append(count)
            
    element_flag = [0]
    element_flag.append(len(answer_list))
    
    i=0
    while(1):
        for index1 in answer_list[element_flag[0]:element_flag[1]]:
            for index2 in answer_list[element_flag[i]:element_flag[i+1]]:
                tmp_list=index1+index2
                count = Contain(db,tmp_list)
                if count>=support and (Contain(answer_list,tmp_list)==0):
                    answer_list.append(tmp_list)
                    answer_support.append(count)

        element_flag.append(len(answer_list))
        i+=1
        if element_flag[i+1] == element_flag[i]:
            break
    Sorting(answer_list,answer_support)
    print(answer_list)
    print(answer_support)
    
    ########  寫檔  ########
    fp = open(output_file, "w")
    for i in range(len(answer_list)):
        fp.write(str(answer_list[i]))
        fp.write(","+str(answer_support[i])+"\n")
    fp.close()
```

```python
import time
start = time.time()

input_file = 'publicdataset_1.csv'
output_file = 'outputdataset_1.txt'
support=2
HW2(support,input_file,output_file)

end = time.time()
elapsed = end - start
print ("Time taken: ", elapsed, "seconds.")
```
